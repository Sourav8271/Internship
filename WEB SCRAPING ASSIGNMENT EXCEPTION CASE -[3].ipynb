```python
# Question 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search
#results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
#scrape all the products available under that product name. Details to be scraped are: "Brand
#Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
#“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 
```

```python
brand=[]
product_name=[]
price=[]
expected_delivery=[]
availability=[]
product_url=[]
```

```python
start=0
end=3
for page in range(start,end):
    
    titles=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-outline"]')
    for i in titles:
        product_url.append(i.get_attribute('href'))
        
    next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-button-accessibility s-pagination-separator"]')
    next_button.click()
    time.sleep(3)
```

```python
for i in product_url:
    driver.get(i)
    time.sleep(5)  # It's generally better to use WebDriverWait instead of fixed sleep time.

    try:
        bd = driver.find_element(By.XPATH, '//*[@id="bylineInfo"]')
        brand.append(bd.text)
    except NoSuchElementException:
        brand.append('-')
        
        
    try:
        rs = driver.find_element(By.XPATH, '//span[@class="a-price-whole"]')
        price.append(rs.text)
    except NoSuchElementException:
        price.append('-')
        
  
    try:
        product_tag = driver.find_element(By.XPATH, '//*[@id="productTitle"]')
        product_name.append(product_tag.text)
    except NoSuchElementException:
        product_name.append('-')
        
        
    try:
        delivery = driver.find_element(By.XPATH, '//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span[1]')
        expected_delivery.append(delivery.text)
    except NoSuchElementException:
        expected_delivery.append('-')
        
    try:
        available = driver.find_element(By.XPATH, '//*[@id="availability"]/span')
        availability.append(available.text)
    except NoSuchElementException:
        availability.append('-')
```
