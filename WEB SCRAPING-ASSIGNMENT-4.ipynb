```python
# Question 4.  Scrape the details of trending repositories on Github.com.
#Url = https://github.com/
#You have to find the following details:
# A) Repository title  B) Repository description  C) Contributors count   D) Language used 
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://github.com/')
```

```python
file_url=[]
title=[]
description=[]
contributors_count=[]
language=[]
```

```python
titles=driver.find_elements(By.XPATH,'//a[@class="Link"]')
for i in titles:
    file_url.append(i.get_attribute('href'))
```

```python
file_url
```

```python
for i in file_url:
    driver.get(i)
  # It's generally better to use WebDriverWait instead of fixed sleep time.

    try:
        ttle= driver.find_element(By.XPATH, '//div[@class=" d-flex flex-wrap flex-items-center wb-break-word f3 text-normal"]')
        title.append(ttle.text)
    except NoSuchElementException:
        title.append('-')
        
        
    try:
        desc= driver.find_element(By.XPATH, '//p[@class="f4 my-3"]')
        description.append(desc.text)
    except NoSuchElementException:
        description.append('-')
        
        
    try:
        count= driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[5]/div/h2/a/span')
        contributors_count.append(count.text)
    except NoSuchElementException:
        contributors_count.append('-')
   


    try:
        lang= driver.find_element(By.XPATH, '/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[6]/div/ul')
        language.append(lang.text)
    except NoSuchElementException:
        language.append('-')
        
        
```

```python
for i in range(len(title)):
    title[i] = title[i].replace('\n', '')
title = [string.replace('\n', '') for string in title]
title
```

```python
description
```

```python
contributors_count
```

```python
for i in range(len(language)):
    language[i] = language[i].replace('\n', '')
language = [string.replace('\n', '') for string in language]
language
```

```python
df=pd.DataFrame({"Title":title,"Description":description,"Contributors Count":contributors_count,"Language":language})
df
```

```python
# Question 8. Details of Datasets from UCI machine learning repositories.
#Url = https://archive.ics.uci.edu/ You have to find the following details:
# A) Dataset name B) Data type C) Task  D) Attribute type  E) No of instances F) No of attribute 
#G) Year 
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
search=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
search.click()
```

```python
dataset_name=[]
attribute=[]
data_type=[]
task=[]
types=[]
instance=[]
features=[]
year=[]
```

```python
url=[]
start=0
end=60
for page in range(start,end):
    
    titles=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a')
    for i in titles:
        url.append(i.get_attribute('href'))
        
    next_button=driver.find_element(By.XPATH,'//button[@class="btn-primary btn-sm btn"][2]')
    next_button.click()
    time.sleep(3)
```

```python
url
```

```python
for i in url:
    driver.get(i)
    # It's generally better to use WebDriverWait instead of fixed sleep time.

    try:
        type= driver.find_element(By.XPATH, '//h1[@class="text-3xl font-semibold text-primary-content"]')
        dataset_name.append(type.text)
    except NoSuchElementException:
        dataset_name.append('-')
        
        
    try:
        type= driver.find_element(By.XPATH, '//p[@class="text-md"]')
        data_type.append(type.text)
    except NoSuchElementException:
        data_type.append('-')
        
  
    try:
        tk= driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        task.append(tk.text)
    except NoSuchElementException:
        task.append('-')
        



    try:
        typ= driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        types.append(typ.text)
    except NoSuchElementException:
        types.append('-')
        
        
    try:
        inst= driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        instance.append(inst.text)
    except NoSuchElementException:
        instance.append('-')
        
        
        
    try:
        yer= driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        year.append(yer.text)
    except NoSuchElementException:
        year.append('-')
        
    try:
        fe= driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        features.append(fe.text)
    except NoSuchElementException:
        features.append('-')
        
```
