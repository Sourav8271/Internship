```python
# Question 1.Write a python program to display all the header tags from wikipedia.org.
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/Wikipedia:Contents/Lists'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Wikipedia=[]
names=soup.find_all('span',class_='mw-headline')
for i in names:
    name=i.text
    Wikipedia.append(name)
print(Wikipedia)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Wikipedia Tags':Wikipedia})
df
```

```python
#Question 2 Write a python program to display IMDB’s Top rated 100 movies’ data i.e. name, rating, year)
```

```python
url='https://www.imdb.com/list/ls006266261/'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
movies_name=[]
names=soup.find_all('div',class_='lister-item-content')
for i in names:
    name=i.h3.a.text
    movies_name.append(name)
movies_name
```

```python
year_of_release=[]
names=soup.find_all('span',class_='lister-item-year text-muted unbold')
for i in names:
    name=i.text.replace('(','').replace(')','').replace('I ','')
    year_of_release.append(name)
year_of_release
```

```python
rating=[]
names=soup.find_all('div',class_='ipl-rating-star small')
for i in names:
    name=i.text.replace('\n','')
    rating.append(name)
rating
```

```python
Duration=[]
names=soup.find_all('span',class_='runtime')
for i in names:
    name=i.text.replace('\n','')
    Duration.append(name)
Duration
```

```python
df=pd.DataFrame({'Movies name':movies_name,'Release year':year_of_release,'Rating':rating,'Movie Duration':Duration})
df
```

```python
#Question 3. Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name 
#ii) Cuisine iii) Location iv) Ratings v) Image URL.
```

```python
url='https://www.dineout.co.in/delhi-restaurants/buffet'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
rest_name=[]
names=soup.find_all('a',class_='restnt-name ellipsis')
for i in names:
    name=i.text
    rest_name.append(name)
rest_name
len(rest_name)
```

```python
dish_name=[]
names=soup.find_all('span',class_='double-line-ellipsis')
for i in names:
    name=i.text
    dish_name.append(name)
dish_name
len(dish_name)
```

```python
loc=[]
rating=[]
names=soup.find_all('div',class_='restnt-loc ellipsis')
for i in names:
    name=i.text
    loc.append(name)
loc
```

```python
for x in range(5):
     names=soup.find_all('div',class_='restnt-rating rating-'+str(x))
print(names)
```

```python
for i in names:
    name=i.text
    rating.append(name)
print(rating)
len(rating)
```

```python
image_url=[]
names=soup.find_all('img',class_='no-img')
for i in names:
    image_url.append(i['data-src'])
image_url
len(image_url)
```

```python
df=pd.DataFrame({'Restaurant Names':rest_name,'Prices|Cuisine':dish_name,'Loacation':loc,'Ratings':rating,'Image Url':image_url})
df
```

```python
#Question 4.Write s python program to display list of respected former finance minister of India(i.e. 
#Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make 
#data frame
```

```python
url='https://presidentofindia.nic.in/former-presidents'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
President_name=[]
names=soup.find_all('div',class_='desc-sec')
for i in names:
    name=i.h3.text
    President_name.append(name)
print(President_name)
len(President_name)
```

```python
Term_of_office=[]
names=soup.find_all('div',class_='desc-sec')
for i in names:
    name=i.h5.text
    Term_of_office.append(name)
print(Term_of_office)
len(Term_of_office)
```

```python
df=pd.DataFrame({'President Name':President_name,'Term of office':Term_of_office})
df
```
